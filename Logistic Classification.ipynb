```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
pd.pandas.set_option('display.max_columns' ,None)
```

```python
df=pd.read_csv("weatherAUS.csv")
```

```python
df.head()
```

```python
df.drop(index=df[df["RainTomorrow"].isnull()==True].index,axis=0,inplace=True)
```

```python
y=df["RainTomorrow"]
```

```python
categ_feat=[ i for i in x.columns if x[i].dtype=='O' and 'Date' not in i]
num_feat=[ i for i in x.columns if x[i].dtype!='O']
categ_nan=[i for i in categ_feat if x[i].isnull().sum()>0]
num_nan=[i for i in num_feat if x[i].isnull().sum()>0]
```

```python
df["Date"].value_counts()
```

```python
df[df["Date"]=='2013-12-01']
```

```python
df["day"]=pd.to_datetime(df["Date"]).dt.day
df["month"]=pd.to_datetime(df["Date"]).dt.month
df["year"]=pd.to_datetime(df["Date"]).dt.year
```

```python
df.head()
```

```python
df.drop(columns=["Date"],inplace=True)
```

```python
x=df.drop(columns=["RainTomorrow"])
```

```python
x.head()
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,train_size=0.8,random_state=0)
```

```python
for i in num_nan:
    medn=x_train[i].median()
    x_train[i]=x_train[i].fillna(medn)
```

```python
print(len(num_feat),"numerical")
print(len(categ_feat),"categorical")
print(len(num_nan),"numerical_nan")
print(len(categ_nan),"catg_nan")
print(len(df.columns))
```

```python
np.percentile(x_train["WindGustSpeed"],25)
```

```python
for i in num_feat:
    q1=np.percentile(x_train[i],25)
    q3=np.percentile(x_train[i],75)
    iqr=q3-q1
    lr=q1-1.5*iqr
    ur=q3+1.5*iqr
    x_train.loc[:,i]=np.where(x_train[i]>ur,ur,x_train[i])
    x_train.loc[:,i]=np.where(x_train[i]<lr,lr,x_train[i])
```

```python
for i in num_nan:
    medn=x_test[i].median()
    x_test[i]=x_test[i].fillna(medn)
```

```python
for i in categ_nan:
    x_test.loc[:,i]=x_test[i].fillna("Missing")
```

```python
for i in num_feat:
    q1=np.percentile(x_test[i],25)
    q3=np.percentile(x_test[i],75)
    iqr=q3-q1
    lr=q1-1.5*iqr
    ur=q3+1.5*iqr
    x_test.loc[:,i]=np.where(x_test[i]>ur,ur,x_test[i])
    x_test.loc[:,i]=np.where(x_test[i]<lr,lr,x_test[i])
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
ls=LogisticRegression(solver='liblinear')
model=ls.fit(x_train1,y_train)
```

```python
y_predict=model.predict(x_test1)
y_predict
```

```python
prob=pd.DataFrame(model.predict_proba(x_test1) ,columns=['Prob of No','Prob of Yes'])
prob
```

```python
comparision=pd.concat([prob,pd.DataFrame(y_predict)],axis=1)
comparision.columns=["No","Yes","O/P"]
```

```python
from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix
from sklearn.metrics import classification_report
```

```python
acc=accuracy_score(y_test,y_predict)
print("Model Testing accuracy score {}" .format(acc))
```

```python
print("Training score {}" .format (ls.score(x_train1,y_train)))
print("Testing score {}" .format (ls.score(x_test1,y_test)))
```

```python
cm=confusion_matrix(y_test,y_predict)
print ("confusion matrix \n" ,cm)
```

```python
print('\nTrue Positives(TP) = ', cm[0,0])

print('\nTrue Negatives(TN) = ', cm[1,1])

print('\nFalse Positives(FP) = ', cm[0,1])

print('\nFalse Negatives(FN) = ', cm[1,0])
```

```python
cm_map=pd.DataFrame(cm, columns=["Actual :1","Actual :0"],index=["Predict :1","Predict :0"])
sns.heatmap(data=cm_map,annot =True , fmt='d', cmap='YlGnBu' )
```

```python
aacuracy_score_fmla=cm[0,0]/(cm[0,0]+cm[0,1]+cm[1,0]+cm[1,1])
print("accoracy score by formula is {0:0.2f}" .format(aacuracy_score_fmla))
```
